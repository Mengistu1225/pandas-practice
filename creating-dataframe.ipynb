```python
import pandas as pd
import numpy as np
```

```python
# creating dataframe using list
data=[["mengistu",25],["hagos",54],['efrem',87],["mahari",27]]
data
df=pd.DataFrame(data,columns=['Name',"Age"])
df

```

```python
# creating dataframe from Dictionaries
data={'name':["jack",'nahom','etsay','haftamu'],
     'Age':[34,65,87,98]}
data
df=pd.DataFrame(data)
df
```

```python
# creating dataframe from tuples
data=(("menge",43),("eyob",30),('yonas',54),('yrga',65))
data
df=pd.DataFrame(data,columns=['Name','Age'])
df
```

```python
# Create a DataFrame with random numbers
df = pd.DataFrame(np.random.randn(5, 3), columns=['A', 'B', 'C'])

# Sample 3 rows randomly
df_sample = df.sample(n=3)

print(df_sample)
```

```python
# simple 2d array into dataframe
arr=np.array([[1,2,3,4,5],[4,5,6,7,8],[9,6,2,7,3]])
print(arr)
df=pd.DataFrame(arr,columns=["A","B","C","D","E"])
df
```

```python
# random data with specific shape
data=np.random.rand(10,5) # 5_rows  and 10_columns
data
df=pd.DataFrame(data,columns=["A","B","C","D","E"])
df
```

```python
# adding indexing and labeles
data=np.arange(10).reshape(5,2)
data
df = pd.DataFrame(data, index=["Row1", "Row2", "Row3", "Row4", "Row5"], columns=["Column1", "Column2"])
df
```

```python
# adding structural from numpy array
data=np.array([(1, 'Alice', 25), (2, 'Bob', 30), (3, 'Charlie', 35)],
                            dtype=[('ID', 'i4'), ('Name', 'U10'), ('Age', 'i4')])
data
df=pd.DataFrame(data)
df
```

```python
# time series data
dates=pd.date_range(start="2023-01-01",periods=10)
dates
data=np.random.rand(10,3)
data
df=pd.DataFrame(data,index=dates,columns=["Feature1", "Feature2", "Feature3"])
df
```
